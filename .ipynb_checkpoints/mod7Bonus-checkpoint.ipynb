```python
# mod7 Bonus

# Kyle Hudson

# Use a different set of time-based data. Import your data and use describe to calculate the basic descriptive statistics for your data set.

import pandas as pd
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()

rec = pd.read_csv('COVID_Vaccinations_Given_to_SF_Residents_by_SF_DPH_Over_Time.csv')

# Use the SciPy stats module linregress function to calculate slope and intercept for the best fit line through the data.

rec.columns = ['Date', 'Cum_Rec']

from scipy import stats

linear_regression = stats.linregress(x=rec.Date, y=rec.Cum_Rec)

print(f'The slope is {linear_regression.slope}')
print(f'The intercept is {linear_regression.intercept}')

# Chart the data with the best-fit line.

import seaborn as sns

sns.set_style('whitegrid')

axes = sns.regplot(x=rec.Date, y=rec.Cum_Rec)

# Use your project to predict a point in the future

int_pred = int(linear_regression.slope * 44918 + linear_regression.intercept)

print(f'The total number of  patients receiving a vaccine by the end of 2022 is {int_pred}')
```
